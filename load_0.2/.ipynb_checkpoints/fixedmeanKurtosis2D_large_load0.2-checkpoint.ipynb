# 2D plot of RT vs CVsq of ST when Kurta = 30 

```python
import pandas as pd
import numpy as np

df = pd.read_csv("DetailsH2_H2_1_fixedmeanKurt_30Kurta_large.csv")
df["Response time"] = df["Response time"].apply(lambda x: float(x.strip('[]')))
#df = df.sort_values(by='mu_kurt')
df.head()
```

```python
filtered_df_30 = df[(df["mu_kurt"] == 30)]

mu_kurt_30 = filtered_df_30["mu_CVsq"]
RT_30 = filtered_df_30["Response time"]
mu_kurt_30 = np.array(mu_kurt_30)*1.5
RT_30 = np.array(RT_30)*1.25

print(filtered_df_30.head())

filtered_df_40 = df[(df["mu_kurt"] == 40)]

mu_kurt_40 = filtered_df_40["mu_CVsq"]
RT_40 = filtered_df_40["Response time"]
mu_kurt_40 = np.array(mu_kurt_40)*1.5
RT_40 = np.array(RT_40)*1.25

print(filtered_df_40.head())

filtered_df_50 = df[(df["mu_kurt"] == 50)]

mu_kurt_50 = filtered_df_50["mu_CVsq"]
RT_50 = filtered_df_50["Response time"]
mu_kurt_50 = np.array(mu_kurt_50)*1.5
RT_50 = np.array(RT_50)*1.25

print(filtered_df_50.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(mu_kurt_30, RT_30, color = 'yellow', label = 'Kurta = 30, Kurts = 40')
plt.scatter(mu_kurt_40, RT_40, color = 'green', label = 'Kurta = 30, Kurts = 30')
plt.scatter(mu_kurt_50, RT_50, color = 'blue', label = 'Kurta = 30, Kurts = 20')
       
plt.xlabel('CV sq of ST curve')
plt.ylabel('RT for H2/H2/1 as per Tarasov for given kurtosis of IAT and ST dists')

plt.xlim(0, 50)
plt.ylim(2, 12)
plt.legend()
plt.show()
```
