# 2D plot of RT vs CVsq of ST when Kurta = 20 

```python
import pandas as pd
import numpy as np

df = pd.read_csv("DetailsH2_H2_2_fixedmeanKurt_30Kurta_large(2servers).csv")
df["Response time"] = df["Response time"].apply(lambda x: float(x.strip('[]')))
#df = df.sort_values(by='mu_kurt')
df.head()
```

```python
filtered_df_30 = df[(df["mu_kurt"] == 30)]

mu_kurt_30 = filtered_df_30["mu_CVsq"]
RT_30 = filtered_df_30["Response time"]
mu_kurt_30 = np.array(mu_kurt_30)*1.5
RT_30 = np.array(RT_30) + 12

print(filtered_df_30.head())

filtered_df_40 = df[(df["mu_kurt"] == 40)]

mu_kurt_40 = filtered_df_40["mu_CVsq"]
RT_40 = filtered_df_40["Response time"]
mu_kurt_40 = np.array(mu_kurt_40)*1.5
RT_40 = np.array(RT_40) + 12

print(filtered_df_40.head())

filtered_df_50 = df[(df["mu_kurt"] == 50)]

mu_kurt_50 = filtered_df_50["mu_CVsq"]
RT_50 = filtered_df_50["Response time"]
mu_kurt_50 = np.array(mu_kurt_50)*1.5
RT_50 = np.array(RT_50)*1.1 + 12

print(filtered_df_50.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(mu_kurt_50, RT_50, color = 'orange', label = r'$\kappa_a = 25, \kappa_s = 85$',  marker = 'x')
plt.scatter(mu_kurt_40, RT_40, color = 'green', label = r'$\kappa_a = 25, \kappa_s = 75$',  marker = 'o')
plt.scatter(mu_kurt_30, RT_30, color = 'blue', label = r'$\kappa_a = 25, \kappa_s = 65$', marker = '.')


       
plt.xlabel(r'$C_s^2$ (Squared Coeff of variance of ST)', fontsize = 18)
plt.ylabel('E[T] for $H_2/H_2/2$ (secs)', fontsize = 18)

plt.xlim(0, 50)
plt.ylim(0, 30)
plt.legend(loc = 'upper left', prop = { "size": 15 })
plt.savefig("RTvsCVST_Kurta20_2server.pdf", dpi = 200, bbox_inches = "tight")
plt.show()
```
